# Importing packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
%matplotlib inline
```

# Reading the cleaned dataset

```python
df = pd.read_csv('medical_insurance_dataset_cleaned.csv')
```

# Viewing the first few rows from the dataset

```python
df.head()
```

Deleting the first column 

```python
df.drop("Unnamed: 0" , axis = 1 , inplace = True)
```

```python
df.head()
```

# Exploratory Data Analysis (EDA) :

Getting the correlation matrix for the dataset

```python
df.corr()
```

Getting the correlation between each Feature and another using Heatmap:

```python
sns.heatmap(df.corr(numeric_only = True) , cmap = 'coolwarm' , annot = False)
plt.show()
```

Implement the regression plot for 'charges' with respect to 'bmi'.

```python
x = df['bmi']
y = df['charges']
sns.regplot(x = 'bmi' , y = 'charges' , data = df)
plt.xlabel("BMI")
plt.ylabel("Charges")
plt.title("Regression Plot for 'bmi' vs. 'charges'")
plt.show()
```

Implement the box plot for 'charges' with respect to 'smoker'.

```python
x = df['smoker'] 
y = df['charges']
sns.boxplot(x = "smoker" , y = "charges" , data = df)
plt.xlabel("Smoker")
plt.ylabel("Charges")
plt.title("Regression Plot for 'Smoker' vs. 'Charges'")
plt.show()
```

Getting the pearson correlation for 'smoker' and 'charges'

```python
pearson_coef_smoker , p_value_smoker = stats.pearsonr(df['smoker'] , df['charges'])
```

```python
print("The Correlation between 'smoker' and 'charges' : ")
print("Pearson Coefficient : " , pearson_coef_smoker)
print("p-value : " , p_value_smoker)
```

```python
0.001 > p_value_smoker
```

from the result we found that :

    1) coefficient is close to 1
    2) p-value = 0 and < (smaller than) 0.001
Then, the correlation between 'smoker' and 'charges' is Large positive Relationship with Strong Certainty in Result.
